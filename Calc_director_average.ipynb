```python
import pandas as pd
```

```python
imdb = pd.read_csv("./infl_adjusted_data.csv")
```

```python
imdb.head()
```

```python
ddata = imdb.loc[:,('director_name','f_adj_gross')]
```

```python
ddata.head()
```

```python
grouped = ddata.groupby('director_name')
```

```python
av = grouped.mean()
```

```python
av.columns=['dir_adj_av_gross']
```

```python
av.head()
```

```python
ddata['dir_adj_av_gross']=0
```

```python
#for each film, look up the director's average earnings and write it to the film data
for index, row in ddata.iterrows():
    thisdir=row["director_name"]
    gr=av.loc[thisdir,'dir_adj_av_gross']
    ddata.loc[index,'dir_adj_av_gross']=gr
    print (index,'\t',thisdir,'\t',gr)
   
```

```python
ddata.head()
```

```python
imdb['dir_adj_av_gross']=ddata.loc[:,'dir_adj_av_gross']
```

```python
imdb=imdb.drop_duplicates()
```

```python
print (imdb.shape)
```

```python
imdb=imdb.dropna()
```
