```python
import pandas as pd

# load imdb csv file into dataframe
imdb = pd.read_csv("./BASE_DATA.csv" )
imdb.head()
```

```python
imdb['ROI_band'] = 0
imdb['profit_level'] = 0
imdb.profit_level = (imdb['gross'] - imdb['budget'])
imdb.head()
```

```python
imdb['ROI_percent'] = 0
imdb.loc[(imdb.profit > 0), 'ROI_percent'] = ( imdb['gross'] / imdb['budget'])* 100
#imdb.ROI_percent = ( imdb['gross'] / imdb['budget'])* 100
imdb.head(20)
```

```python
successfulMovies = imdb[(imdb.profit > 0)]
successfulMovies.ROI_percent.describe()
```

```python
%matplotlib inline
successfulMovies.plot.scatter(x='budget', y='ROI_percent')
```

```python
highestROI = imdb[(imdb.ROI_percent > 10000)]
highestROI
```
