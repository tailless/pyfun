```python
import pandas as pd
```

```python
imdb = pd.read_csv("./movie_rate_adjusted_metadata.csv")
```

```python
imdb.shape
```

```python
#check if any n/a data left
imdb = imdb.dropna()
```

```python
imdb.shape
```

```python
#check if any duplicates
imdb = imdb.drop_duplicates()
```

```python
imdb.shape
```

```python
#check what columns we've got
imdb.head()
```

```python
#read in US CPI data
cpi=pd.read_csv("./uscpi.csv", index_col=0, header=0)
```

```python
#check read in ok
cpi.head()
```

```python
#add columns to the imdb dataframe - for adjusted film gross earnings and adjusted budget
imdb['f_adj_gross']=0
imdb['adj_budget']=0
#check columns added
imdb.head()
```

```python
#calculate the inflation adjusted film earnings and budget and store in the new columns
for index, row in imdb.iterrows():
    thisyr=row['title_year']
    thiscpi=cpi.loc[thisyr,'factor']
    
    thisgr=row['gross']
    newgr=thisgr*thiscpi
    
    thisbudg=row['budget']
    newbudg=thisbudg*thiscpi
    
    #use this to check what's been calculated
    #print (thisyr,'\t', thiscpi, '\t', thisgr,'\t', newgr,'\t', thisbudg,'\t', newbudg )
    
    #write the new data to the new imdb columns
    imdb.loc[index,'f_adj_gross']=newgr
    imdb.loc[index,'adj_budget']=newbudg
```

```python
imdb
```

```python
#extract director and film adjusted gross data
ddata = imdb.loc[:,('director_name','f_adj_gross')]
    
```

```python
df = pd.DataFrame(ddata)
```

```python
#check we've got what we want
df.head()
```

```python
#create grouped object by director name
grouped = df.groupby('director_name')
```

```python
#calculate average for each director name group
av = grouped.mean()
```

```python
#change column label so we know we now have the director's adjusted average gross earnings
av.columns=['dir_adj_av_gross']
```

```python
av.head()
```

```python
mean = pd.DataFrame(av)
```

```python
mean.head()
```

```python
mean.columns
```

```python
#add a column for director average earnings
df['dir_adj_av_gross']=0
```

```python
df.head()
```

```python
#for each film, look up the director's average earnings and write it to the film data
for index, row in df.iterrows():
    thisdir=row["director_name"]
    gr=mean.loc[thisdir,'dir_adj_av_gross']
    df.loc[index,'dir_adj_av_gross']=gr
    print (index,'\t',thisdir,'\t',gr)
   
         
```

```python
#write the new data to the original database
imdb['dir_adj_av_gross']=df.loc[:,'dir_adj_av_gross']
```
