```python
import pandas as pd

# load imdb csv file into dataframe
imdb = pd.read_csv("./movies_with_ROI.csv" )
imdb.head()
```

```python
dfGenres = imdb.loc[:, 'genres']
dfGenres.head()
```

```python
gArray = dfGenres.str.split('|', expand=False)
gArray
```

```python
allGenres = gArray.sum()
allGenres
```

```python
import numpy as np
uniqueGenres = np.unique(allGenres)
uniqueGenres
```

```python
uniqueGenres.shape
```

```python
dfSplitDummies = dfGenres.str.get_dummies(sep='|')
dfSplitDummies
```

```python
imdb.shape
```

```python
moviesWithSplitGenres = pd.concat([imdb, dfSplitDummies], axis=1, join_axes=[imdb.index])
moviesWithSplitGenres
```

```python
moviesWithSplitGenres.shape
```

```python
# Write ROI dataframe to csv
moviesWithSplitGenres.to_csv("./movies_with_SplitGenres.csv", index=False, encoding="utf-8")
```
